```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
print(flights.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Aggregate data to get total passengers per year
flights_yearly = flights.groupby('year', as_index=False)['passengers'].sum()

# Create the line chart
plt.plot(flights_yearly['year'], flights_yearly['passengers'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Total Passengers per Year')

# Show the plot
plt.show()
```

```python
# 2. Scatter Plot: Total Passenger Volume by Month (1949-1960)
# Group by month and sum the passengers
monthly_passengers = flights.groupby('month', as_index=False, observed=True)['passengers'].sum()

# Create the scatter plot
plt.scatter(monthly_passengers['month'], monthly_passengers['passengers'])

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total Passengers')
plt.title('Total Passenger Volume per Month (1949-1960)')

# Show the plot
plt.show()
```

```python
# 3. Horizontal Stacked Bar Chart: Total Number of Passengers by Month and Year

# Group the data by month and year, and then sum the number of passengers
monthly_passengers_by_year = flights.groupby (['month', 'year'], observed=False)['passengers'].sum().unstack()

# Create the horizontal stacked bar chart
monthly_passengers_by_year.plot(kind='barh', stacked=True)

# Add labels and title
plt.xlabel('Total Number of Passengers')
plt.ylabel('Month')
plt.title('Total Number of Passengers by Month and Year')

# Move the legend outside the plot
plt.legend(title='Year', bbox_to_anchor=(1.2, 1), loc='upper right')

# Show the plot
plt.show()
```

```python
# 4. Vertical Stacked Bar Chart: Total Number of Passengers by Month and Year

# Group the data by month and year
monthly_passengers_by_year = flights.groupby(['month', 'year'], observed=False)['passengers'].sum().unstack()

# Create the vertical stacked bar chart
monthly_passengers_by_year.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Total Number of Passengers by Month and Year')

# Move the legend outside the plot
plt.legend(title='Year', bbox_to_anchor=(1.2, 1), loc='upper right')

# Show the plot
plt.show()
```
